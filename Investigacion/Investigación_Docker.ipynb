**Introducción a Docker**

Docker es una aplicación de código abierto que nos permite desplegar aplicaciones autocontenidas. Esto significa que por ejemplo, una aplicación que necesite tener Python instalado en el sistema para ejecutarse, al ser autocontenida, significa que ya lo incluye en el “container”, y esto permite ejecutarlo en cualquier máquina, al igual que cualquier otro programa que se necesite para ejecutarlo. Esto hace que se pueda ejecutar en cualquier sistema operativo sin que aparezcan los problemas típicos de falta de archivos, programas, etc. O sea que lleva en un contenedor el programa principal junto con todos los programas que se necesiten para ejecutarlo.



**Diferencia entre Docker y una máquina virtual**

Una máquina virtual contiene un sistema operativo completo, con todos sus archivos, drivers de hardware, incluso archivos que no necesitaremos para la ejecución del programa, e inclusive muchos archivos que la máquina o sistema operativo base ya lo tengan, lo que hace que ocupe muchos recursos innecesarios, más aún si queremos ejecutar varias aplicaciones.

La alternativa es el uso de “contenedores”. Lo que se hace es que los ficheros y configuraciones se compartan con el sistema operativo base (virtualización a nivel sistema operativo). Los contenedores son agrupaciones de procesos que están aislados, lo llevo a cualquier lugar y funcionan en cualquier lado.

**Funcionamiento de Docker**

Docker va a correr un contenedor a partir de una imagen

Imagen004.png

El **OS** es la distribución del sistema operativo(Ubuntu, Fedora, etc) , no el kernel, el kernel se compartirá con el host. 
El **software** van a ser los programas, los paquetes, las librerías (por ejemplo las librerías Xarray, Datacube, MatplotLib, etc) y todas las dependencias que permiten que el programa se ejecute.
La **app** es el código de nuestra aplicación principal.
Después corremos esa imagen en el contenedor, ¿cómo? Con Dockerfile


Imagen005.png


Las imágenes se generan usando un **dockerfile**, que es un archivo con una serie de instrucciones que indican cómo crear una imagen. Después con el comando **dockerbuild** se genera la imagen, y para correr esa imagen usamos **dockerrun** y se ejecuta el contenedor. 


**Instalación de Docker**
Entrar a https://www.docker.com/products/docker-desktop y elegir la que corresponda a nuestro SO e instalarlo.
Registrarse en Docker bajo sencillos pasos para poder hacer uso del repositorio de Docker.

Controlar que Docker fue instalado correctamente en nuestro sistema. En la terminal de comandos poner docker -v, con eso nos tendría que salir la versión de Docker instalada.
Entrar a Docker Hub https://hub.docker.com/ en la cual estarán los repositorios de los programas. Tomar en cuenta que están las imágenes de aplicaciones originales y también otras aplicaciones subidas por usuarios con modificaciones, complementos, o inclusive packs completos con varias aplicaciones en una sola imagen etc. 
Al entrar a la imagen que busquemos en Docker Hub, a la derecha nos aparecerá el comando para instalarla.

**Principales comandos Docker**

**docker pull nombreImagen** (nos descarga la imagen del repositorio Docker) (se le agrega :numVersión si queremos una versión anterior)

**docker run nombreImagen** (ejecuta determinada imagen y la convierte en contenedor, lo bueno es que si no la tenemos instalada, nos las descargará automáticamente y después la ejecutará.)
Ojo, esto no significa que tengamos dicha imagen/programa instalado en nuestra máquina, sino que lo ejecuta como si fuese una máquina virtual.

**docker images** (nos muestra todas las imágenes que tenemos descargadas en nuestra máquina)
docker ps nos dice qué contenedores se están ejecutando



**¿Cómo creamos un contenedor?**

Creamos un Dockerfile, que en resumen son las instrucciones que debe seguir Docker para construir un contenedor
vim Dockerfile

La primera línea dice **FROM**, ahí tenemos que poner una imagen padre, que tiene todo lo que necesitamos (por ejemplo si vamos a correr una aplicación Node, pondremos la imagen node que tiene todas las dependencias que necesitaremos.
Por ejemplo: **FROM node:12.22.1-alpine3.11**

La segunda línea **WORKDIR** es donde declaramos nuestra carpeta de trabajo, o sea que todos los comandos que se especifiquen en el archivo se correrán en esa carpeta
Por ejemplo: **WORKDIR /app**

La tercera línea **COPY . .** (punto espacio punto) copia todo el directorio de nuestra aplicación en la que estamos trabajando al contenedor que está en /app

La cuarta línea **RUN** compila todo lo que tenemos en nuestro código
Por ejemplo **RUN yarn install --production**

La línea **CMD** especifica el comando que queremos que corra 
Por ejemplo **CMD [“node”, “/app/src/index.js”] **especifica que corra el comando node y le pase como argumento, o sea que cuando corra esta imagen correrá el binario /app/src/index.js
