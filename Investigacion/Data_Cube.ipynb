# Investigación de software y librerías de Open Data Cube

Open Data Cube es un software de gestión y análisis de datos geoespaciales de código abierto (los cuales se obtienen a partir de los satélites Landsat, Sentinel, MODIS, etc). Es un conjunto de bibliotecas de Python y una base de datos PostgreSQL que ayudan a trabajar con datos ráster geoespaciales.
Página del proyecto: [opendatacube.org](https://opendatacube.org)

# Conceptos
Pero primero entendamos algunos conceptos...
Un **ráster** es una matriz de filas y columnas en la que cada celda representa un valor que representa información, como temperatura, elevación, etc (datos continuos) o también datos en los que se pueden diferenciar diferentes objetos, como un lago, un bosque, una ruta, etc. (datos discretos)

Un **Data Cube** es un “cubo” de datos, o sea, una matriz de múltiples dimensiones en donde encontraremos una gran cantidad de datos, y cada dimensión es un tipo de dato en particular (por ejemplo latitud, longitud, elevación, etc).
